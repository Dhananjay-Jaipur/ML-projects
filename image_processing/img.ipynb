## Take input of targeted image.

```python
target = i.imread("white.png")
plt.imshow(target)
```

```python
target
```

## Extract colour.

```python
img1_hsv = cv2.cvtColor(img1, cv2.COLOR_RGB2HSV)
plt.imshow(img1_hsv)
```

```python
#Convert the images to the HSV color space (hue, saturation, value). This conversion allows us to easily isolate the dark colors based on their value
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
target_hsv = cv2.cvtColor(target, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_hsv)
```

```python
img
```

```python
img_hsv
```

```python
plt.imshow(target_hsv)
```

```python
#Define the lower and upper threshold values for the dark colors you want to extract. In the HSV color space, the value ranges from 0 to 255
lower_threshold = np.array([0, 0, 0])
upper_threshold = np.array([75, 50, 50]) # 10/20
```

```python
#color mask isolates a particular color in an image
#Create a mask using the threshold values to isolate the dark colors in the source image.
dark_mask = cv2.inRange(img, lower_threshold, upper_threshold)

#dark_mask = cv2.inRange(img_hsv, lower_threshold, upper_threshold)
```

```python
plt.imshow(dark_mask)
```

## Apply the mask to the target image

Bitwise AND (cv2.bitwise_and): Performs a bitwise AND operation on two input images. It sets each pixel in the output image to 255 (1 in binary) if the corresponding pixel values in both input images are non-zero; otherwise, it sets the output pixel value to 0.

Bitwise OR (cv2.bitwise_or): Performs a bitwise OR operation on two input images. It sets each pixel in the output image to 255 (1 in binary) if the corresponding pixel values in either of the input images are non-zero; otherwise, it sets the output pixel value to 0.

Bitwise XOR (cv2.bitwise_xor): Performs a bitwise XOR operation on two input images. It sets each pixel in the output image to 255 (1 in binary) if the corresponding pixel values in the input images differ; otherwise, it sets the output pixel value to 0.

Bitwise NOT (cv2.bitwise_not): Performs a bitwise NOT operation on a single input image. It inverts the pixel values of the image, setting each 0 pixel to 255 and vice versa.


```python
#Replace the corresponding pixels in the destination image with the extracted dark colors:

target_dark = cv2.bitwise_or(target, dark_mask)
```

```python
plt.imshow(target_dark)
```

```python
#Convert the target image back to the BGR color space
target_dark_bgr = cv2.cvtColor(target_dark, cv2.COLOR_HSV2BGR)
```

```python
# Display or save the result
plt.imshow('Target Image with Dark Colors', target_dark_bgr
```
