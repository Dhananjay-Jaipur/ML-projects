```python
import cv2 
import matplotlib.pyplot as plt


# reading image 
img = cv2.imread("jaipur.jpg") 

# Convert image to grayscale 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  

# Apply blur  
gray = cv2.medianBlur(gray, 5) 

# Apply edge detection
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9) 

# Apply bilateral filter for smoothing image
color = cv2.bilateralFilter(img, 90, 100, 250) 

#cartoonify(result)
# Combine smoothed image with edges
cartoon = cv2.bitwise_and(color, color, mask=edges) 

plt.imshow(img)
plt.show()
plt.imshow(edges) 
plt.show()
plt.imshow(cartoon) 
plt.show()

cv2.imwrite("Cartoon.jpg", cartoon)
```

adaptiveThreshold - creates a binary image where pixels are set to white if they are above a certain threshold(255), otherwise black

Edges - are defined as a set of connected pixels that form a boundary between two disjoint regions. They are also known as boundaries and are identified based on sharp changes in image brightness.
